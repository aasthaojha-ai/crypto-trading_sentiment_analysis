# Crypto Trading Data Analysis
### This project analyzes crypto trading data along with Fear & Greed Index sentiment data using Python.

Two datasets are used in this project:
- Historical crypto trader data
- Bitcoin Fear & Greed sentiment data

The main tasks performed in this notebook are:

- Data cleaning
- Handling missing values
- Feature engineering
- Merging datasets
- Exploratory Data Analysis (EDA)
- Data visualization using Matplotlib, Seaborn, and Plotly

The analysis helps understand trader behavior, profit/loss patterns, leverage usage, and how market sentiment affects trading activity.

## Stage 1: Importing all the necessary Libraries

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

## Stage 2: Reading csv files using Pandas

```python
fgi_raw = pd.read_csv('fear_greed_index.csv')
hist_raw = pd.read_csv('historical_data.csv')

print("Fear & Greed Index Raw Shape:", fgi_raw.shape)
print("Historical Trading Data Raw Shape:",hist_raw.shape)

#display first few rows of the both dataset
display(fgi_raw.head())
display(hist_raw.head())
```

## Stage 3: Finding missing values

```python
missing_fgi = fgi_raw.isnull().sum()
missing_hist = hist_raw.isnull().sum()

print("Missing values in Fear & Greed dataset:")
display(missing_fgi[missing_fgi > 0])

print("Missing values in Historical Trading dataset:")
display(missing_hist[missing_hist > 0])
```

## Stage 4: Data Cleaning

```python
# making copies of datasets
fgi_clean = fgi_raw.copy()
hist_clean = hist_raw.copy()

# checking duplicates
print("Duplicate rows in Fear & Greed dataset:", fgi_clean.duplicated().sum())
print("Duplicate rows in Historical dataset:", hist_clean.duplicated().sum())

# removing duplicates
fgi_clean = fgi_clean.drop_duplicates()
hist_clean = hist_clean.drop_duplicates()

# removing missing values
fgi_clean = fgi_clean.dropna()
hist_clean = hist_clean.dropna()

print("Cleaning completed")
```

## Stage 5: Date and Time Conversion

```python
# converting date columns
fgi_clean['date'] = pd.to_datetime(fgi_clean['date'])
hist_clean['Timestamp IST'] = pd.to_datetime(hist_clean['Timestamp IST'])

print("Date conversion completed")
```

 ## Stage 6: C hecking Invalid Values

```python
# checking invalid values

print("Invalid values in Fear & Greed dataset:")
print(fgi_clean[fgi_clean['value'] < 0].shape[0])

print("Invalid values in Historical dataset:")
print(hist_clean[hist_clean['Execution Price'] <= 0].shape[0])

# removing invalid values

hist_clean = hist_clean[hist_clean['Execution Price'] > 0]

print("Invalid values removed")
```

 ## Stage 7: Feature Engineering

```python
# feature engineering

# extracting trade date
hist_clean['trade_date'] = hist_clean['Timestamp IST'].dt.date

# creating profit/loss category
hist_clean['pnl_category'] = np.where(
    hist_clean['Closed PnL'] > 0,
    'Profit',
    'Loss'
)

# calculating trade volume
hist_clean['trade_volume'] = (
    hist_clean['Execution Price'] * hist_clean['Size Tokens']
)

print("Feature engineering completed")
```

## Stage 8: Saving Cleaned Datasets

```python
# saving cleaned datasets

fgi_clean.to_csv("fear_greed_cleaned.csv", index=False)
hist_clean.to_csv("historical_cleaned.csv", index=False)

print("Files saved successfully")
```

## Stage 9: Merging the Datasets

```python
# merging both datasets using date column

merged_df = pd.merge(
    hist_clean,
    fgi_clean[['date', 'value', 'classification']],
    left_on='trade_date',
    right_on='date',
    how='inner'
)

# renaming columns
merged_df.rename(columns={'value': 'fgi_value'}, inplace=True)
print("Datasets merged successfully")

display(
    merged_df[
        ['trade_date', 'Coin', 'fgi_value', 'classification', 'Closed PnL']
    ].head()
)
```

## Stage 10: Exploratory Data Analysis (EDA)

We will now analyze trader activity and market sentiment using different visualizations created with Matplotlib, Seaborn, and Plotly.

### Visualization 1: Sentiment vs Average PnL

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=merged_df,
    x='classification',
    y='Closed PnL'
)

plt.title("Sentiment vs Average PnL")
plt.xlabel("Market Sentiment")
plt.ylabel("Closed PnL")

plt.show()
```

### Observation
Average profit and loss changes across different market sentiment categories. Some sentiment conditions show higher profitability compared to others.

### Visualization 3: Trade Count by Market Sentiment

```python
trade_counts = merged_df['classification'].value_counts()

plt.figure(figsize=(8,5))

sns.barplot(
    x=trade_counts.index,
    y=trade_counts.values
)

plt.title("Trade Count by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Number of Trades")

plt.show()
```

### Observation
The chart shows how trading activity changes across different market sentiment categories. Some sentiment conditions have higher trade frequency compared to others.

### Visualization 4: ## Buy vs Sell PnL Analysis

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=merged_df,
    x='classification',
    y='Closed PnL',
    hue='Side'
)

plt.title("Buy vs Sell PnL Analysis")
plt.xlabel("Market Sentiment")
plt.ylabel("Closed PnL")

plt.show()
```

### Observation
The chart compares the profit and loss values for Buy and Sell trades across different market sentiment categories.

### Visualization 5: Profit and Loss Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(
    merged_df['Closed PnL'],
    bins=30,
    kde=True
)

plt.title("Profit and Loss Distribution")
plt.xlabel("Closed PnL")
plt.ylabel("Frequency")

plt.show()
```

### Observation
The histogram shows the distribution of profit and loss values across trades. Most trades are concentrated around smaller PnL values, while some trades have larger profits or losses.

### Visualization 6: Cumulative PnL Trend

```python
# calculating cumulative pnl by date

daily_stats = merged_df.groupby('trade_date')['Closed PnL'].sum().reset_index()

daily_stats['Cumulative PnL'] = daily_stats['Closed PnL'].cumsum()

# interactive line chart

fig = px.line(
    daily_stats,
    x='trade_date',
    y='Cumulative PnL',
    title='Cumulative PnL Over Time'
)

fig.show()
```

### Observation
The chart shows how cumulative profit and loss changes over time based on daily trading activity.

### Visualization 8: Trading Activity Over Time

```python
# counting trades by date

daily_activity = merged_df.groupby('trade_date').size().reset_index(name='Trade Count')

plt.figure(figsize=(8,5))

sns.lineplot(
    data=daily_activity,
    x='trade_date',
    y='Trade Count'
)

plt.title("Trading Activity Over Time")
plt.xlabel("Date")
plt.ylabel("Trade Count")

plt.show()
```

### Observation
The line chart shows how trading activity changes over time based on the number of trades executed each day.

### Visualization 9: Correlation Heatmap

```python
plt.figure(figsize=(8,6))

numeric_cols = [
    'fgi_value',
    'Closed PnL',
    'Execution Price',
    'Size Tokens',
    'trade_volume'
]

corr_matrix = merged_df[numeric_cols].corr()

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

### Observation
The heatmap shows the relationship between different numerical features such as trade volume, execution price, and profit/loss.

### Visualization 9: Correlation Pairplot

```python
sns.pairplot(
    merged_df[
        ['fgi_value', 'Closed PnL', 'trade_volume']
    ]
)

plt.show()
```

### Observation
The pairplot helps visualize relationships and distributions between market sentiment, trade volume, and profit/loss values. It shows how these numerical features vary and interact with each other.
